```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,100)
```

```python
def retsin(input):
    return np.sin(input)
```

```python
def retcos(input):
    return np.cos(input)
```

```python
figure, plots = plt.subplots(2)
plots[0].plot(x, retsin(x))
plots[1].plot(x, retcos(x))
figure.savefig("hi.pdf")
```
